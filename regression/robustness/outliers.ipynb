```python
def produceOutlier(n):
    np.random.seed(123)
    X_out = np.random.uniform(-10, 10, n).reshape(n,1)
    y_out = np.abs(np.random.normal(0, 20, n).reshape(n,1))
    return X_out, y_out
```

```python
def regressionLoss(y_pred, y_true, metric = 'RAE', outlier_filter = False, outlier_thres = 20):
    metric = metric.lower()
    if outlier_filter:
        std = regressionLoss(y_pred, y_true, 'RMSE', False)
        outlier_flag = (np.abs(y_pred - y_true) > std*outlier_thres)
        num_outlier = np.sum(outlier_flag)
        return regressionLoss(y_pred[~outlier_flag], y_true[~outlier_flag], metric, False), num_outlier

    y_true, y_pred = np.array(y_true), np.array(y_pred)
    y_mean = y_true.mean()
    if metric == 'rae':
        return sum(abs(y_pred - y_true)) / sum(abs(y_mean - y_true))
    elif metric == 'rse':
        return sum((y_pred - y_true)**2) / sum((y_mean - y_true)**2)
    elif metric == 'rmse':
        return ((y_pred - y_true)**2).mean()**.5
    elif metric == 'r square':
        sst = sum((y_true - y_mean) ** 2)
        ssr = sum((y_true - y_pred) ** 2)
        return 1 - ssr/sst
```

```python
import time, os, json
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib as mpl
from operator import is_not
from functools import partial
%matplotlib inline
```

```python
class MedianKNNRegressor(KNeighborsRegressor):
    def predict(self, X):


        neigh_dist, neigh_ind = self.kneighbors(X)

        weights = _get_weights(neigh_dist, self.weights)

        _y = self._y
        if _y.ndim == 1:
            _y = _y.reshape((-1, 1))

        ######## Begin modification
        if weights is None:
            y_pred = np.median(_y[neigh_ind], axis=1)
        else:
            # y_pred = weighted_median(_y[neigh_ind], weights, axis=1)
            raise NotImplementedError("weighted median")
        ######### End modification

        if self._y.ndim == 1:
            y_pred = y_pred.ravel()

        return y_pred    
```

```python
title_list = ['(1) linear function', '(2) quadratic function', '(3) exponential function',
            '(4) cosine function', '(5) l1-norm function', '(6) piecewise function']
```

```python
width = 25
height = 5

plt.rc('font', family='serif', serif='times new roman')
plt.rc('text', usetex=True)
plt.rc('xtick', labelsize=28)
plt.rc('ytick', labelsize=28)
plt.rc('axes', labelsize=28)
plt.rc('axes', linewidth=1)
mpl.rcParams['patch.linewidth']=0.5 #width of the boundary of legend

fig, ax = plt.subplots(nrows=1, ncols=6, sharex=True, sharey=False) #plot two subgraphs in one figure, 1 by 2 size, sharing y-axis
fig.subplots_adjust(left=.06, bottom=.17, right=0.99, top=0.9, wspace=0.25) #margin of the figure
fig.set_size_inches(width, height) #exact size of the figure


func_list = ['linear', 'quadratic', 'exponential', 'cosine', 'l1norm', 'piecewise']

for i in range(6):
    func = func_list[i]
    y, e, gptj_y, gptj_e = [], [], [], []
    knn_y, knn_e, knn3_y, knn3_e = [], [], [], []
    for n in n_list:
        dir_name = 'out_n_%d' % n
        rmse_list = np.zeros(3)
        gptj_rmse = np.zeros(3)
        knn_rmse = np.zeros(3)
        knn3_rmse = np.zeros(3)
        for sim_idx in range(1, 4):
            file = '%s/data_%d/%s_n_250_p_1_int_0_(-10.0,10.0)_noise_0.10_all.json' % (dir_name, sim_idx, func)
            # print(file)
            with open(file, 'r') as f:
                 data_json = json.load(f)
            rmse_list[sim_idx-1] = regressionLoss(data_json['gpt3_test_y'], data_json['test_y'])
            gptj_rmse[sim_idx-1] = regressionLoss(data_json['gptj_test_y'], data_json['test_y'])
            knn = MedianKNNRegressor(n_neighbors=5)
            knn.fit(data_json['train_x'], data_json['train_y'])
            knn_test_y = knn.predict(data_json['test_x'])
            knn_rmse[sim_idx-1] = regressionLoss(knn_test_y, data_json['test_y'])
            
            knn3 = MedianKNNRegressor(n_neighbors=3)
            knn3.fit(data_json['train_x'], data_json['train_y'])
            knn3_test_y = knn3.predict(data_json['test_x'])
            knn3_rmse[sim_idx-1] = regressionLoss(knn3_test_y, data_json['test_y'])
        y.append(rmse_list.mean())
        e.append(rmse_list.std())
        knn_y.append(knn_rmse.mean())
        knn_e.append(knn_rmse.std())
        knn3_y.append(knn3_rmse.mean())
        knn3_e.append(knn3_rmse.std())
        gptj_y.append(gptj_rmse.mean())
        gptj_e.append(gptj_rmse.std())
    # ax[i].errorbar(pc_list, knn_y, e)
    ax[i].errorbar(pc_list, gptj_y, gptj_e, label = 'LIFT/GPT-J', linewidth = 2)
    ax[i].errorbar(pc_list, y, e, label = 'LIFT/GPT-3', linewidth = 2)
    ax[i].errorbar(pc_list, knn_y, knn_e, label = 'Median 5NN', linewidth = 2)
    ax[i].errorbar(pc_list, knn3_y, knn3_e, label = 'Median 3NN', linewidth = 2)
    # ax[i].set_xticks(pc_list)
    ax[i].set_title(title_list[i], fontsize = 28)
ax[0].legend(fontsize = 20)
fig.supxlabel('Percentage of Outliers',fontsize = 28)
fig.supylabel('RAE',fontsize = 28)
plt.savefig('../figures/outliers.pdf')
None
```
